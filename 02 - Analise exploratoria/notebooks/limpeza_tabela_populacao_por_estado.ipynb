Criar coluna com a regi√£o, relacionada a cada estado:
'CENTRO OESTE', 'NORDESTE', 'NORTE', 'SUDESTE', 'SUL'

```python
df_populacao_por_estado["regiao"] = "NORDESTE"

df_populacao_por_estado.loc[
    (df_populacao_por_estado['sigla_uf'] == 'RS') | 
    (df_populacao_por_estado['sigla_uf'] == 'SC') | 
    (df_populacao_por_estado['sigla_uf'] == 'PR'),'regiao'] = "SUL"

df_populacao_por_estado.loc[
    (df_populacao_por_estado['sigla_uf'] == 'SP') | 
    (df_populacao_por_estado['sigla_uf'] == 'RJ') | 
    (df_populacao_por_estado['sigla_uf'] == 'MG') | 
    (df_populacao_por_estado['sigla_uf'] == 'ES'),'regiao'] = "SUDESTE"

df_populacao_por_estado.loc[
    (df_populacao_por_estado['sigla_uf'] == 'MT') | 
    (df_populacao_por_estado['sigla_uf'] == 'MS') | 
    (df_populacao_por_estado['sigla_uf'] == 'GO'),'regiao'] = "CENTRO OESTE"

df_populacao_por_estado.loc[
    (df_populacao_por_estado['sigla_uf'] == 'AM') | 
    (df_populacao_por_estado['sigla_uf'] == 'RR') | 
    (df_populacao_por_estado['sigla_uf'] == 'AP') | 
    (df_populacao_por_estado['sigla_uf'] == 'TO') | 
    (df_populacao_por_estado['sigla_uf'] == 'RO') | 
    (df_populacao_por_estado['sigla_uf'] == 'AC') | 
    (df_populacao_por_estado['sigla_uf'] == 'PA'),'regiao'] = "NORTE"


```
