```python
import pandas as pd
import numpy as np

df_produto_interno_estados = pd.read_csv('../data/tabelas antes da limpeza/tabela_produto_interno_estados.csv', delimiter=";")
```

Lendo o dataset de PIB, e colocando a coluna Unidade da Federação como index

```python
df_produto_interno_estados = df_produto_interno_estados.set_index(df_produto_interno_estados["Unidade da Federação"])
df_produto_interno_estados[:10]
```

Fazendo o melt do dataset, para ter o ano e pib como colunas

```python
df_produto_interno_estados = pd.melt(df_produto_interno_estados, id_vars=['Unidade da Federação'], 
var_name='ano', value_name='pib')
```

```python
df_produto_interno_estados = df_produto_interno_estados.rename(columns={'Unidade da Federação': "sigla_uf"})
```

Converter nome do Estado para Sigla

```python
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Rondônia", "RO", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Acre", "AC", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Amazonas", "AM", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Roraima", "RR", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Pará", "PA", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Amapá", "AP", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Tocantins", "TO", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Maranhão", "MA", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Piauí", "PI", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Ceará", "CE", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Rio Grande do Norte", "RN", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Paraíba", "PB", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Pernambuco", "PE", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Alagoas", "AL", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Sergipe", "SE", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Bahia", "BA", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Minas Gerais", "MG", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Espírito Santo", "ES", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Rio de Janeiro", "RJ", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("São Paulo", "SP", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Paraná", "PR", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Santa Catarina", "SC", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Rio Grande do Sul", "RS", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Mato Grosso do Sul", "MS", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Mato Grosso", "MT", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Goiás", "GO", regex=True)
df_produto_interno_estados["sigla_uf"] = df_produto_interno_estados["sigla_uf"].replace("Distrito Federal", "DF", regex=True)
```

```python
df_produto_interno_estados["sigla_uf"].unique()
```

```python
df_produto_interno_estados[:10]
```

Criar coluna com a região, relacionada a cada estado:
'CENTRO OESTE', 'NORDESTE', 'NORTE', 'SUDESTE', 'SUL'

```python
df_produto_interno_estados["regiao"] = "NORDESTE"

df_produto_interno_estados.loc[
    (df_produto_interno_estados['sigla_uf'] == 'RS') | 
    (df_produto_interno_estados['sigla_uf'] == 'SC') | 
    (df_produto_interno_estados['sigla_uf'] == 'PR'),'regiao'] = "SUL"

df_produto_interno_estados.loc[
    (df_produto_interno_estados['sigla_uf'] == 'SP') | 
    (df_produto_interno_estados['sigla_uf'] == 'RJ') | 
    (df_produto_interno_estados['sigla_uf'] == 'MG') | 
    (df_produto_interno_estados['sigla_uf'] == 'ES'),'regiao'] = "SUDESTE"

df_produto_interno_estados.loc[
    (df_produto_interno_estados['sigla_uf'] == 'MT') | 
    (df_produto_interno_estados['sigla_uf'] == 'MS') | 
    (df_produto_interno_estados['sigla_uf'] == 'GO'),'regiao'] = "CENTRO OESTE"

df_produto_interno_estados.loc[
    (df_produto_interno_estados['sigla_uf'] == 'AM') | 
    (df_produto_interno_estados['sigla_uf'] == 'RR') | 
    (df_produto_interno_estados['sigla_uf'] == 'AP') | 
    (df_produto_interno_estados['sigla_uf'] == 'TO') | 
    (df_produto_interno_estados['sigla_uf'] == 'RO') | 
    (df_produto_interno_estados['sigla_uf'] == 'AC') | 
    (df_produto_interno_estados['sigla_uf'] == 'PA'),'regiao'] = "NORTE"


```
